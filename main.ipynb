**Data Preparation**

```python
import pandas as pd

df = pd.read_csv('/content/Pandas-/walmart_weekly_sales.csv')


#data types
#df.dtypes

#inspect all columns
df.columns.values
```

**Drop unnamed columns**

```python
df = df.loc[:,~df.columns.str.contains('^Unnamed')]

df.head(5)
```

**Check for missing values in the new data set**

```python
df.isnull().sum()
```

**Check the dataTypes**

```python
df.dtypes
```

**Describe the dataframe**

```python
df.describe()
```

```python



max_sales_row = df.loc[df['Weekly_Sales'].idxmax()]

store_with_max_sales = max_sales_row['Store']

print(store_with_max_sales)
```

```python
# Check if it was a holiday week
was_holiday = max_sales_row['Holiday_Flag'] == 1

print(was_holiday)
```

```python
# Extract the month from the Date column
date_of_max_sales = pd.to_datetime(max_sales_row['Date'])
month_of_max_sales = date_of_max_sales.strftime('%B')
print(f"The month of the week with the highest sales is {month_of_max_sales}")
```

```python
# Get the sales amount for that week
sales_amount = max_sales_row['Weekly_Sales']
print(f"The sales amount for that week was ${sales_amount:.2f}")
```

```python
df['Weekly_Sales_Increased'] = df['Weekly_Sales'] * 1.07

# Display the updated DataFrame
df.head()
```

```python
df.head()
```

```python

# Calculate costs as new columns
df['Labour_Cost'] = df['Weekly_Sales'] * 0.286
df['Product_Cost'] = df['Weekly_Sales'] * 0.39
df['Maintenance_Cost'] = df['Weekly_Sales'] * 0.124

# Calculate profit as a new column
df['Profit'] = df['Weekly_Sales'] - (df['Labour_Cost'] + df['Product_Cost'] + df['Maintenance_Cost'])

# Display the updated DataFrame
df.head()
```
